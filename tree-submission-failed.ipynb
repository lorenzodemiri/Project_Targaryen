```python
train = pd.read_csv('/kaggle/input/jane-street-market-prediction/train.csv')
```

```python
train = train.query('date > 85').reset_index(drop = True) 
train = train[train['weight'] != 0]
```

```python
train.fillna(0, inplace=True)
train.isna().sum()
```

```python
train['action'] = ((train['resp_1'] > 0).astype('int'))
train['action']
```

```python
y= train['action']
y
```

```python
train = train.drop(['action'], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

import pickle

model = DecisionTreeClassifier(random_state=777)

model.fit(train, y)

pkl_filename = "/kaggle/working/pickle_tree_model.pkl"
with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)
```

```python
import pickle
#load our saved model 
Pkl_Filename = "./pickle_tree_model.pkl"
with open(Pkl_Filename, 'rb') as file:  
    model = pickle.load(file)
```

```python
import janestreet
env = janestreet.make_env() # initialize the environment
iter_test = env.iter_test() # an iterator which loops over the test set

for (test_df, sample_prediction_df) in iter_test:
    test_df.fillna(0)
    sample_prediction_df.action = model.predict(test_df) #make your 0/1 prediction here
    env.predict(sample_prediction_df)
```
